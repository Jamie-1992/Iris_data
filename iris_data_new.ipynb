# K-nearest neighbours algorithm

## Data preparation

```python
#remember we saw that petal length and width was easiest way to seperate the species 

features = iris[['petal length (cm)', 'petal width (cm)']]

target = iris['species']

from sklearn.model_selection import train_test_split

feat_train, feat_test, targ_train, targ_test = train_test_split(features, target,
                                                               test_size = 0.3,
                                                               random_state = 1,
                                                               stratify = target) #stratify makes sure distribution of labels between train and and test is the same

print(targ_train.value_counts()) #same proportion of each target class
print(targ_test.value_counts()) #same proportion of each target class


#note if stratify not on then some classes in the train/test samples will be more or less than other classes.
#what it does is makes sure it is proportional to the whole dataset. in  this case there should be 33.33% for each class.





```

## Modeling

```python
#Now time to build and train the knn model

from sklearn.neighbors import KNeighborsClassifier

#creating an instance of the knn class

knn = KNeighborsClassifier(n_neighbors = 5) #the hyperparameter is how many neighbours to use before using the model

#training the model 

knn.fit(feat_train, targ_train) #used default metric which is minkowski distance and p = 2. p means power so squared. 
                                # sum(|x - y|^p)^(1/p) 
    
```

# Prediction 

## Label prediction 

```python
targ_pred = knn.predict(feat_test) #Note uses the model we fitted with training data to predict labels with the features
print(targ_pred[:5])  ## Note: predicting the coded integers for each species 
                 # 2 is iris-virginica, 0 is iris-setosa, 1 is iris-versicolor

```

## Probability prediction 

```python
targ_pred_prob = knn.predict_proba(feat_test)
print(targ_pred_prob[10:12]) #this includes 11th and 12th flower. remember indexing starts at 0 so thats why 10 is 11 and doesn't include 12 which is 13
#for the 11th flower observation, it is predicted 100% it is iris-setosa
# for the 12 flower observation, it is predicted 80% chance of being iris-virginica

#What it tells us is that of the five nearest neighbours of the 12th flower in the testing set, 
#1 is an iris-versicolor, the other 4 are iris-virginica. 

print(targ_pred[10:12]) #this is a hard prediction where it just uses the labels whereas the prob one outputs probability of an observation belonging to a class (soft)
```

# Model evaluation 

## Accuracy

```python
#accuracy is the most straighforward metric in classification problems 

print((targ_pred== targ_test).sum()) #classifier made one error. 44/45. this adds up all the trues which = 1 and 1 was false which = 0


print((targ_pred==targ_test).sum()/targ_test.size) #this ends up being 44/45 because size is 45
#print((targ_pred==targ_test.values).sum()/targ_test.size) #this outputs the same as previous line. 

#or can do it this way 

knn.score(feat_test, targ_test)
print(feat_test)
print(targ_test)
```

## Confusion matrix

```python
#Classification accuracy alone can be misleading if there is an unequal number of observations in each class or if there 
#are more than two classes in the dataset. Calculating a confusion matrix will provide a better idea of what the 
#classification is getting right and what types of errors it is making.

from sklearn.metrics import confusion_matrix

confusion_matrix(targ_test, targ_pred, labels = [0,1,2]) #remember 0 = iris-setosa, 1 is iris-versicolor and 2 is iris-versicolor

#can visualize doing

from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(knn, feat_test, targ_test, #confusion matrix takes two arguments since we are using knn model
                     cmap = plt.cm.Blues);
```

## K-fold cross validation 


```python
from sklearn.model_selection import cross_val_score
#Creating a new KNN model
knn_cv = KNeighborsClassifier(n_neighbors = 3)

#fitting the 3 nearest neighbor model
cv_scores = cross_val_score(knn_cv, features, target, cv=5) #5 fold validation, each holdout set has 20% of the original data

#remember for each prediction, we use 80% of data in train and 20% in test and do this 5 times.
#print each cv score (accuracy)
print(cv_scores) #accuracy fluctuates around since not the exact same data is in the holdout each time.

#We average the scores to get overall accuracy 

cv_scores.mean()
```

## Grid Search 

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# create new knn model

knn2 = KNeighborsClassifier()
#create a dict of all values we want to test for n_neighbors

param_grid = {'n_neighbors': np.arange(2,10)} #numpy array will create numbers from 2 to 9 with a stepsize of 1 as default.

# use gridsearch to test all values for n_neighbors

knn_gscv = GridSearchCV(knn2, param_grid, cv=5)

#fit model to data 

knn_gscv.fit(features,target)

#to check top n_neighbors value
knn_gscv.best_params_ 

#what is the accuracy when neighbors = 4

knn_gscv.best_score_ # finding optimal hyperparameter improved model by 1%

#Now to build final model 

knn_final = KNeighborsClassifier(n_neighbors=knn_gscv.best_params_['n_neighbors']) #indexing n_neighbors makes sure we print out 4 instead of
#the dictionary key with the value 4

#print(knn_gscv.best_params_) this would do dict with value

knn_final.fit(features,target)
targ_pred = knn_final.predict(features)
knn_final.score(features,target) #predicts iris species with 97% accuracy




```
